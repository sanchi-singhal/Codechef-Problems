### MEAN AND MEDAIN (MEANMEDIAN)

https://www.codechef.com/START55D/problems/MEANMEDIAN

#### Approach
from equation (p+q+r)/3=x<br>
we have given q=y and x is also given<br>
we need to find p and q.<br>
trying fixing the value of one of p and q in various cases as shown below:


```python
t=int(input())
for i in range(t):
    x,y=map(int,input().split())
    if(x==y):
        p=q=r=x
    else:
        q=y
        if(q<0):
            if(x<y):
                r=0
                p=3*x-y
            else:
                p=q-1
                r=3*x-p-y
        elif(q>=0):
            if(x>y):
                p=0
                r=3*x-y
            else:
                r=q+1
                p=3*x-y-r
    print(p,q,r)
```
