### BALL AND BOXES(BALLBOX)

### Approach
First fullfill the required condition(satisfy given constraints)<br>
Thenn,check for the boundary answer (in this case minimum number of balls)<br>

```python
# cook your dish here
import math
t=int(input())
for i in range(t):
    n,k=map(int,input().split())
    # l=0
    # for  j in range(1,k+1):             # approach 1,2 (not optimized solution)
    #     l=l+j
    # if(n>=l):
    #     print("YES")
    # else:
    #     print("NO")
    if(k%2!=0):                          #approach 3 (optimized solution)
        l=math.ceil(k/2)
        req=k*l
    else:
        l=math.ceil((k+1)/2)
        req=(k+1)*l-(k+1)
    if(n>=req):
        print("YES")
    else:
        print("NO")
```
